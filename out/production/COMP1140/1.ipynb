# Code Walk Q&A - 中英对照

## 变量追踪 (Variable Tracking)

**Q: Trace through someFunction(6). What are x and y after each step?**
**问：追踪 someFunction(6) 的执行。每步之后 x 和 y 的值是什么？**

**A:** Initial: x=6. After y=x+1: y=7. After x=y+2: x=9. After y=x-3: y=6. After x=y*2: x=12. After y=x-(y+1): y=5. Returns 17.
**答：** 初始：x=6。y=x+1 后：y=7。x=y+2 后：x=9。y=x-3 后：y=6。x=y*2 后：x=12。y=x-(y+1) 后：y=5。返回 17。

## 数据结构定义 (Data Structure Definitions)

**Q: Is BinaryTree a self-referential data definition?**
**问：BinaryTree 是自引用数据定义吗？**

**A:** Yes. The Node record contains two BinaryTree fields (left, right), so the type refers to itself in its own definition.
**答：** 是的。Node 记录包含两个 BinaryTree 字段（left, right），所以类型在自己的定义中引用了自己。

**Q: What's the template for processing BinaryTree?**
**问：处理 BinaryTree 的模板是什么？**

**A:** Switch on tree cases: Leaf() returns base case, Node(left, right) makes recursive calls on both left and right subtrees.
**答：** 对树进行 switch：Leaf() 返回基础情况，Node(left, right) 对左右子树进行递归调用。

## 折叠函数 (Fold Functions)

**Q: What's the difference between Fold and FoldLeft execution?**
**问：Fold 和 FoldLeft 的执行有什么区别？**

**A:** Fold processes elements in natural order like A+(B+(C+start)). FoldLeft in this library reverses element order, producing different results for non-commutative operations.
**答：** Fold 按自然顺序处理元素，如 A+(B+(C+start))。这个库中的 FoldLeft 会逆转元素顺序，对非交换操作产生不同结果。

**Q: Which parameter is the accumulator in (x,y) -> x+y?**
**问：在 (x,y) -> x+y 中哪个参数是累加器？**

**A:** y is the accumulator. x is the current list element, y is the intermediate result being accumulated.
**答：** y 是累加器。x 是当前列表元素，y 是正在累积的中间结果。

## 递归和互相递归 (Recursion and Mutual Recursion)

**Q: What is mutual recursion?**
**问：什么是互相递归？**

**A:** When two functions call each other. Function f calls g, and g calls f, creating indirect recursive calls through the other function.
**答：** 两个函数相互调用。函数 f 调用 g，g 调用 f，通过另一个函数创建间接递归调用。

**Q: How do you count nodes in a BinaryTree?**
**问：如何计算 BinaryTree 中的节点数？**

**A:** Switch on tree: Leaf() returns 0, Node(left, right) returns 1 + count(left) + count(right).
**答：** 对树进行 switch：Leaf() 返回 0，Node(left, right) 返回 1 + count(left) + count(right)。

## 可变状态 (Mutable State)

**Q: What's the difference between HashMap Put and ConsList Map Put?**
**问：HashMap 的 Put 和 ConsList Map 的 Put 有什么区别？**

**A:** HashMap Put modifies the map in-place (void return). ConsList Map Put returns a new map, leaving the original unchanged (immutable).
**答：** HashMap Put 原地修改映射（返回 void）。ConsList Map Put 返回新映射，保持原始映射不变（不可变）。

**Q: When would you choose HashMap over ConsList Map?**
**问：什么时候选择 HashMap 而不是 ConsList Map？**

**A:** Choose HashMap for performance-critical applications with frequent updates. Choose ConsList Map for functional programming style and thread safety.
**答：** 在频繁更新的性能关键应用中选择 HashMap。在函数式编程风格和线程安全需求中选择 ConsList Map。

## 循环 (Loops)

**Q: How do you ensure a while loop terminates?**
**问：如何确保 while 循环终止？**

**A:** Ensure the loop condition eventually becomes false by modifying variables that affect the condition inside the loop body.
**答：** 通过在循环体内修改影响条件的变量，确保循环条件最终变为 false。

**Q: What's the difference between enhanced for and regular for?**
**问：增强型 for 和常规 for 的区别是什么？**

**A:** Enhanced for iterates over collections automatically. Regular for uses explicit index control with initialization, condition, and increment.
**答：** 增强型 for 自动遍历集合。常规 for 使用显式的索引控制，包括初始化、条件和递增。

## 设计策略 (Design Strategy)

**Q: When do you use structural recursion vs iteration?**
**问：什么时候使用结构化递归而不是迭代？**

**A:** Use structural recursion for self-referential data structures. Use iteration for counted loops or when performance is critical.
**答：** 对自引用数据结构使用结构化递归。对计数循环或性能关键时使用迭代。
